```python
import pandas as pd

# Load dataset
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# View top rows
print(df.head())
```

```python
# Filter data for India
india = df[df['Country Name'] == 'India']
print(india)
```

```python
import matplotlib.pyplot as plt

# Extract year columns
years = india.columns[4:-1]  # All years (skip metadata)
population = india.loc[:, years].values.flatten()

# Plot
plt.figure(figsize=(12,6))
plt.plot(years, population, marker='o', linestyle='-', color='royalblue')
plt.title("India's Population Growth (1960â€“2022)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
selected_years = ['1960', '1980', '2000', '2020', '2022']
selected_pop = india[selected_years].values.flatten()

plt.bar(selected_years, selected_pop, color='skyblue')
plt.title("India's Population at Selected Years")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
